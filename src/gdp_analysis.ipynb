```python
df = pd.read_csv("data/wdi_data (5).csv")
df.head()
```

```python
df.drop(columns=["Unnamed: 0"],inplace=True)
df
```

## Separating The Countries Selected from the complete Data

```python
df_ind = df[df["Country Name"]=='India']
df_ind.head()
```

```python
df_usa = df[df["Country Name"]=='United States']
df_usa.head()
```

```python
df_chi = df[df["Country Name"]=='China']
df_chi.head()
```

```python
df_uk = df[df["Country Name"]=='United Kingdom']
df_uk.head()
```

```python
df_aus = df[df["Country Name"]=='Australia']
df_aus.head()
```

```python
df_uae = df[df["Country Name"]=='United Arab Emirates']
df_uae.head()
```

## Adding all the selected Countries in a separate Dataframe

```python
frames=[df_aus,df_chi,df_ind,df_uae,df_uk,df_usa]
result= pd.concat(frames)
result
```

```python
df_2015 = result.query("Year == 2015") ## Now separating the Year 2015 from the data frame
df_2015.sample(10)
```

```python
df_2015 = df_2015.query("indicator in ['NY.GDP.PCAP.PP.CD','SP.POP.TOTL','GB.XPD.RSDV.GD.ZS','IP.JRN.ARTC.SC','TX.VAL.TECH.MF.ZS']"	)
df_2015.head()
```

```python
df_pivot = df_2015.pivot(index =['Country Code','Country Name','Region','Income Group','Lending Type'], columns='indicator', values='value').reset_index()
df_pivot.head()
```

**GDP Per Capita of different countries:**
- GDP Per Capita is a measurement of the GDP per person in a countryâ€™s population. It indicates that the amount of output or income per person in an economy can indicate average productivity or average living standards. 

```python
fig = px.pie(
    df_pivot,
    values='NY.GDP.PCAP.PP.CD', 
    names='Country Name',
    title='TOP 6 Countries by 2015 GDP per capita,PPP(Current International $)',
    hover_name='Country Name',
    
)
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.show()
```

## Population Of Selected Countries

```python
fig = px.bar(
    df_pivot,
    x="Country Name",
    y="SP.POP.TOTL",
    title="TOP 6 Countries By 2015 Population,Total",
    color='Country Name',
    labels={"SP.POP.TOTL":"2015 Population,Total"}
).update_xaxes(categoryorder="total descending")
fig.show()
```

# Effect of  Science& Technology on GDP:
As mentioned earlier GDP is calculated with the countries economy. This means if any country spends on the development of the technology it would rather develop the GDP as well.
- Spending on technology would help in the raise of the below factors:
  - **Workforce Transformation**: This means outsourcing the work with some companies which would help in the development of small businesses which would rather directly help in the growth of the countryâ€™s GDP.
  - **Investment in Startups**: If a country spends some money on developing and implementing the ideas of startups to a larger extent the technology would rise and would help in GDP growth.
  - **Research & Development**: If a product is deployed for the consumerâ€™s use which is developed by a country and being exported to different countries. It would play a vital role in the growth of the GDP of that country. 

```python
fig = px.bar(
    df_pivot,
    x="Country Name",
    y="GB.XPD.RSDV.GD.ZS",
    title="TOP 6 Countries by 2015 Research and Development Expenditure(% of GDP)",
    color='Country Name',
    labels={"GB.XPD.RSDV.GD.ZS":"2015 Research and Development Expenditure(% of GDP)"}
    
).update_xaxes(categoryorder="total descending")
fig.show()
```

```python
fig = px.line(df_pivot2,
    x="Year",
    y="TX.VAL.TECH.MF.ZS",
    title="TOP 6 Countries by High-technology exports (% of manufactured exports)",
    color='Country Name',
    labels={"Year":"Year","TX.VAL.TECH.MF.ZS":"High-technology exports (% of manufactured exports)"},
    hover_name='Country Name',
    
)
fig.show()
```

# Scientific and Technical Journal Articles:

```python
fig = px.scatter(
    df_pivot.dropna(),
    x="NY.GDP.PCAP.PP.CD",
    y="IP.JRN.ARTC.SC",
    title="TOP 6 Countries by 2015 Scientific and Technical Journal Articles",
    labels={"NY.GDP.PCAP.PP.CD":"2015 GDP per Capita,PPP(current international $)","IP.JRN.ARTC.SC":"2015 Scientific and Technical Journal Articles"},
    color='Country Name',
    hover_name='Country Name',
    size="SP.POP.TOTL",
    text="Country Name",
    size_max=75,
    height =800
)
fig.update_layout(showlegend=False)
fig.update_traces(textposition="top center")
fig.show()
```
